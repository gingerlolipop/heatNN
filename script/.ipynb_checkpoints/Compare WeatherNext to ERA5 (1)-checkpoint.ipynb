## Plot BC maps for July 1st

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np
from matplotlib.patches import Circle
import pandas as pd

# Download the July ERA5 data file from Google Cloud Storage
!gsutil cp gs://gencast/ERA5_2021_07/data_stream-oper_stepType-instant.nc /tmp/era5_july.nc

# Open the July dataset
ds_july = xr.open_dataset('/tmp/era5_july.nc', engine='netcdf4')
print("July Dataset info:")
print(ds_july)

# Select the first four time steps (assuming valid_time is the time coordinate)
ds_july_first4 = ds_july.isel(valid_time=slice(0, 4))

# Define coordinates for the two locations:
vancouver_coord = {'latitude': 49.28, 'longitude': -123.12}
lytton_coord   = {'latitude': 50.68, 'longitude': -121.95}

# Create a 2x2 grid plot for the four time steps.
fig, axes = plt.subplots(2, 2, figsize=(18, 12),
                         subplot_kw={'projection': ccrs.PlateCarree()})
axes = axes.flatten()

# Set contour levels (here from -5°C to 40°C)
levels = np.arange(-5, 41, 2)

for ax, t in zip(axes, ds_july_first4.valid_time.values):
    # Select the time slice and convert from Kelvin to Celsius
    t2m_slice = ds_july_first4['t2m'].sel(valid_time=t, method='nearest') - 273.15
    
    # Plot the data on the map
    im = t2m_slice.plot(ax=ax,
                        transform=ccrs.PlateCarree(),
                        cmap='RdYlBu_r',
                        levels=levels,
                        extend='both',
                        add_colorbar=False)
    
    # Add map features
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.STATES, linestyle=':')
    ax.set_extent([-139, -113, 48, 60], crs=ccrs.PlateCarree())
    ax.gridlines(draw_labels=True)
    ax.set_title(f"2m Temp at {np.datetime_as_string(t, unit='h')} UTC", fontsize=10)
    
    # Overlay circles for Vancouver and Lytton
    # Adjust radius (in degrees) as needed; here 0.5° is used as an example.
    vancouver_circle = Circle((vancouver_coord['longitude'], vancouver_coord['latitude']),
                                radius=0.5,
                                edgecolor='blue',
                                facecolor='none',
                                linewidth=2,
                                transform=ccrs.PlateCarree())
    lytton_circle = Circle((lytton_coord['longitude'], lytton_coord['latitude']),
                           radius=0.5,
                           edgecolor='red',
                           facecolor='none',
                           linewidth=2,
                           transform=ccrs.PlateCarree())
    ax.add_patch(vancouver_circle)
    ax.add_patch(lytton_circle)

# Adjust the subplots to leave space on the right for the colorbar
fig.subplots_adjust(right=0.85)

# Add a separate axis for the colorbar on the right
cbar_ax = fig.add_axes([0.88, 0.15, 0.03, 0.7])
cbar = fig.colorbar(im, cax=cbar_ax, label='Temperature (°C)')
cbar.ax.tick_params(labelsize=8)

plt.suptitle('ERA5 2m Temperature – First 4 Times from July Dataset (UTC)', fontsize=16)
plt.tight_layout(rect=[0, 0, 0.85, 0.95])
plt.show()

# ----------------------------------------------------
# Create a summary table of statistics for Vancouver and Lytton
# ----------------------------------------------------
# Extract temperature data (Kelvin converted to °C) at the two locations from the first four time steps
vancouver_temps = ds_july_first4['t2m'].sel(latitude=vancouver_coord['latitude'],
                                             longitude=vancouver_coord['longitude'],
                                             method='nearest') - 273.15
lytton_temps = ds_july_first4['t2m'].sel(latitude=lytton_coord['latitude'],
                                           longitude=lytton_coord['longitude'],
                                           method='nearest') - 273.15

# Compute summary statistics
summary_data = {
    "Location": ["Vancouver (49.28, -123.12)", "Lytton (50.68, -121.95)"],
    "Mean (°C)": [float(vancouver_temps.mean()), float(lytton_temps.mean())],
    "Min (°C)": [float(vancouver_temps.min()), float(lytton_temps.min())],
    "Max (°C)": [float(vancouver_temps.max()), float(lytton_temps.max())],
    "Std (°C)": [float(vancouver_temps.std()), float(lytton_temps.std())]
}

df_summary = pd.DataFrame(summary_data)
print("\nSummary Statistics for Selected Locations (First 4 Times):")
print(df_summary)
```

## Plot time-series for Vancouver and Lytton

```python
## @title Plot time-series for Vancouver and Lytton

import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Download both June and July ERA5 data files
!gsutil cp gs://gencast/ERA5_2021_06/data_stream-oper_stepType-instant.nc /tmp/era5_june.nc
!gsutil cp gs://gencast/ERA5_2021_07/data_stream-oper_stepType-instant.nc /tmp/era5_july.nc

# Open the datasets
ds_june = xr.open_dataset('/tmp/era5_june.nc', engine='netcdf4')
ds_july = xr.open_dataset('/tmp/era5_july.nc', engine='netcdf4')

print("June Dataset info:")
print(ds_june)
print("July Dataset info:")
print(ds_july)

# Concatenate the two datasets along the time (valid_time) dimension.
ds_combined = xr.concat([ds_june, ds_july], dim="valid_time")
print("\nCombined Dataset info:")
print(ds_combined)

# Define coordinates for the two locations:
# Vancouver: (latitude: 49.28, longitude: -123.12)
# Lytton:    (latitude: 50.68, longitude: -121.95)
vancouver_coord = {'latitude': 49.28, 'longitude': -123.12}
lytton_coord   = {'latitude': 50.68, 'longitude': -121.95}

# Extract the time series at these locations and convert from Kelvin to Celsius
vancouver_da = ds_combined['t2m'].sel(latitude=vancouver_coord['latitude'],
                                      longitude=vancouver_coord['longitude'],
                                      method='nearest') - 273.15
lytton_da   = ds_combined['t2m'].sel(latitude=lytton_coord['latitude'],
                                      longitude=lytton_coord['longitude'],
                                      method='nearest') - 273.15

# Convert the DataArrays to pandas Series using valid_time as the index
vancouver_series = pd.Series(vancouver_da.values, 
                             index=pd.to_datetime(vancouver_da.valid_time.values))
lytton_series = pd.Series(lytton_da.values, 
                           index=pd.to_datetime(lytton_da.valid_time.values))

# Compute daily averages and maximums by grouping by the date portion of the index
vancouver_daily_mean = vancouver_series.groupby(vancouver_series.index.date).mean()
vancouver_daily_max  = vancouver_series.groupby(vancouver_series.index.date).max()

lytton_daily_mean = lytton_series.groupby(lytton_series.index.date).mean()
lytton_daily_max  = lytton_series.groupby(lytton_series.index.date).max()

# Plot the daily mean and maximum temperatures with consistent colors per location
plt.figure(figsize=(14, 7))
# Vancouver in blue
plt.plot(vancouver_daily_mean.index, vancouver_daily_mean, color='blue', marker='o', linestyle='-', 
         label='Vancouver (49.28, -123.12) Daily Mean')
plt.plot(vancouver_daily_max.index, vancouver_daily_max, color='blue', marker='^', linestyle='--', 
         label='Vancouver (49.28, -123.12) Daily Max')
# Lytton in red
plt.plot(lytton_daily_mean.index, lytton_daily_mean, color='red', marker='s', linestyle='-', 
         label='Lytton (50.68, -121.95) Daily Mean')
plt.plot(lytton_daily_max.index, lytton_daily_max, color='red', marker='D', linestyle='--', 
         label='Lytton (50.68, -121.95) Daily Max')

plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Daily Mean and Maximum 2m Temperature (June & July Combined, UTC)')
plt.grid(True)
plt.legend()
plt.gcf().autofmt_xdate()  # Auto-format the x-axis dates for clarity
plt.tight_layout()
plt.show()

# Optional: Print the extracted time series values for verification
print("Vancouver time series (°C):", vancouver_series.values)
print("Lytton time series (°C):", lytton_series.values)

# Close the datasets when finished
ds_june.close()
ds_july.close()
```

## Load all CSV results from BigQuery

```python
import subprocess
import pandas as pd
import geopandas as gpd
from shapely import wkt

# ---------------------------------------
# 1) CONFIGURATION
# ---------------------------------------
load_all = True  
# ---------------------------------------
# 2) LIST & LOAD CSV FILES
# ---------------------------------------
cmd = ['gsutil', 'ls', 'gs://gencast/weathernext/weathernext_bc_heatwave_2021_avg/*.csv']
file_list = subprocess.check_output(cmd).decode('utf-8').strip().split('\n')

# Optionally limit to just the first N files if load_all is False
if not load_all:
    file_list = file_list[:2]

print("Files to load:", file_list)

df_list = []
for f in file_list:
    df_temp = pd.read_csv(f, names=column_names, header=None)
    df_list.append(df_temp)

# Concatenate into one DataFrame
df_all = pd.concat(df_list, ignore_index=True)

# ---------------------------------------
# 3) CONVERT TO GEODATAFRAME
# ---------------------------------------
# Convert times to datetime
df_all['init_time'] = pd.to_datetime(df_all['init_time'], errors='coerce')
df_all['forecast_time'] = pd.to_datetime(df_all['forecast_time'], errors='coerce')

# Convert WKT polygons to geometry
df_all['geometry'] = df_all['geography_polygon_text'].apply(wkt.loads)

# Create a GeoDataFrame in EPSG:4326 (lat/lon)
gdf_wn = gpd.GeoDataFrame(df_all, geometry='geometry', crs='EPSG:4326')

# ---------------------------------------
# 4) QUICK INSPECTION
# ---------------------------------------
print("\nCombined GeoDataFrame:")
print(gdf_wn.head())
print("\nEarliest forecast_time:", gdf_wn['forecast_time'].min())
print("Latest forecast_time:", gdf_wn['forecast_time'].max())
```

## Plot BC maps for July 1st

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib as mpl
import pandas as pd

# --- Filter for July 1, 2021 ---
start_time = pd.to_datetime("2021-07-01 00:00:00+00:00")
end_time   = pd.to_datetime("2021-07-02 00:00:00+00:00")
july1 = gdf_wn[(gdf_wn['forecast_time'] >= start_time) & (gdf_wn['forecast_time'] < end_time)].copy()

# --- Convert 2m temperature from Kelvin to Celsius ---
july1['avg_2m_temperature_c'] = july1['avg_2m_temperature'] - 273.15

# --- Define target times ---
target_times = pd.to_datetime([
    "2021-07-01 00:00:00+00:00",
    "2021-07-01 06:00:00+00:00",
    "2021-07-01 12:00:00+00:00",
    "2021-07-01 18:00:00+00:00"
])

# --- Set up 2x2 grid of maps ---
fig, axs = plt.subplots(2, 2, figsize=(16, 12), subplot_kw={'projection': ccrs.PlateCarree()})
axs = axs.flatten()

# Use a consistent temperature scale: from -5°C to 40°C
vmin, vmax = -5, 40

for ax, t in zip(axs, target_times):
    subset = july1[july1['forecast_time'] == t]
    if subset.empty:
        ax.text(0.5, 0.5, f"No data for {t}", transform=ax.transAxes, ha='center')
    else:
        subset.plot(ax=ax, column='avg_2m_temperature_c', cmap='RdYlBu_r',
                    vmin=vmin, vmax=vmax, edgecolor='none')
    for feat in [cfeature.COASTLINE, cfeature.BORDERS, cfeature.STATES]:
        ax.add_feature(feat, linestyle=':')
    ax.set_extent([-139, -113, 48, 60], crs=ccrs.PlateCarree())
    ax.set_title(t.strftime('%Y-%m-%d %H:%M UTC'), fontsize=10)

# Adjust subplots to leave space on the right for the colorbar
fig.subplots_adjust(right=0.85)

# --- Shared colorbar ---
norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
sm = mpl.cm.ScalarMappable(norm=norm, cmap='RdYlBu_r')
sm._A = []
cbar_ax = fig.add_axes([0.88, 0.15, 0.03, 0.7])
cbar = fig.colorbar(sm, cax=cbar_ax, label="2m Temperature (°C)")
cbar.ax.tick_params(labelsize=10)

plt.suptitle("WeatherNext 2m Temperature – July 1, 2021 (All Polygons)", fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import shapely.geometry as geom

# --- 1. Filter for June 25 to July 5 forecast times ---
start_time = pd.to_datetime("2021-06-25 00:00:00+00:00")
end_time   = pd.to_datetime("2021-07-05 23:59:59+00:00")
wn_period = gdf_wn[(gdf_wn['forecast_time'] >= start_time) & (gdf_wn['forecast_time'] <= end_time)].copy()

# --- 2. Convert temperature from Kelvin to Celsius ---
wn_period['avg_2m_temperature_c'] = wn_period['avg_2m_temperature'] - 273.15

# --- 3. Define locations and select nearby polygons ---
# Define points for Vancouver and Lytton
vancouver_point = geom.Point(-123.12, 49.28)
lytton_point   = geom.Point(-121.95, 50.68)
buffer_radius = 0.1  # degrees; adjust as needed

# Filter polygons that lie within the buffer of each point
vancouver_data = wn_period[wn_period['geometry'].distance(vancouver_point) < buffer_radius]
lytton_data   = wn_period[wn_period['geometry'].distance(lytton_point) < buffer_radius]

# --- 4. Compute daily mean and maximum temperatures ---
# Create a new column for the forecast date
vancouver_data['forecast_date'] = vancouver_data['forecast_time'].dt.date
lytton_data['forecast_date'] = lytton_data['forecast_time'].dt.date

vancouver_daily_mean = vancouver_data.groupby('forecast_date')['avg_2m_temperature_c'].mean()
vancouver_daily_max  = vancouver_data.groupby('forecast_date')['avg_2m_temperature_c'].max()

lytton_daily_mean = lytton_data.groupby('forecast_date')['avg_2m_temperature_c'].mean()
lytton_daily_max  = lytton_data.groupby('forecast_date')['avg_2m_temperature_c'].max()

# --- 5. Plot the time series ---
plt.figure(figsize=(14, 7))
# Vancouver in blue
plt.plot(vancouver_daily_mean.index, vancouver_daily_mean, marker='o', linestyle='-', color='blue', label='Vancouver Daily Mean')
plt.plot(vancouver_daily_max.index, vancouver_daily_max, marker='^', linestyle='--', color='blue', label='Vancouver Daily Max')
# Lytton in red
plt.plot(lytton_daily_mean.index, lytton_daily_mean, marker='s', linestyle='-', color='red', label='Lytton Daily Mean')
plt.plot(lytton_daily_max.index, lytton_daily_max, marker='D', linestyle='--', color='red', label='Lytton Daily Max')

plt.xlabel("Date")
plt.ylabel("2m Temperature (°C)")
plt.title("Daily Mean and Max 2m Temperature (WeatherNext)\nVancouver vs Lytton: June 25 to July 5, 2021 (UTC)")
plt.legend()
plt.grid(True)
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
import subprocess
import pandas as pd
import geopandas as gpd
from shapely import wkt
from shapely.geometry import box
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib as mpl
import numpy as np
from matplotlib.patches import Circle

# ====================================================
# 1) Load CSV Files and Build a Combined DataFrame
# ====================================================
# List all CSV files in the folder using gsutil
cmd = ['gsutil', 'ls', 'gs://gencast/weathernext/weathernext_bc_heatwave_2021_avg/*.csv']
file_list = subprocess.check_output(cmd).decode('utf-8').strip().split('\n')

# Optionally limit to a few files for testing (e.g., first 2 files)
load_all = True
if not load_all:
    file_list = file_list[:2]
print("Files to load:", file_list)

# Define column names (confirmed that no data loss occurs)
column_names = [
    "init_time",
    "forecast_time",
    "geography_polygon_text",
    "geography_text",
    "avg_total_precipitation_12hr",
    "avg_100m_u_component_of_wind",
    "avg_100m_v_component_of_wind",
    "avg_10m_u_component_of_wind",
    "avg_10m_v_component_of_wind",
    "avg_2m_temperature",
    "avg_mean_sea_level_pressure",
    "avg_sea_surface_temperature",
]

# Read and concatenate CSV files
df_list = []
for f in file_list:
    df_temp = pd.read_csv(f, names=column_names, header=None)
    df_list.append(df_temp)
df_all = pd.concat(df_list, ignore_index=True)

# ====================================================
# 2) Convert to GeoDataFrame and Check Lead Times
# ====================================================
# Convert time columns to datetime (timezone-aware)
df_all['init_time'] = pd.to_datetime(df_all['init_time'], errors='coerce')
df_all['forecast_time'] = pd.to_datetime(df_all['forecast_time'], errors='coerce')

# Convert the WKT polygon text to geometry
df_all['geometry'] = df_all['geography_polygon_text'].apply(wkt.loads)

# Create a GeoDataFrame in EPSG:4326 (lat/lon)
gdf_wn = gpd.GeoDataFrame(df_all, geometry='geometry', crs='EPSG:4326')

# Compute lead time in days
gdf_wn['lead_time_days'] = (gdf_wn['forecast_time'] - gdf_wn['init_time']).dt.days

# Quick inspection
print("\nCombined GeoDataFrame:")
print(gdf_wn.head())
print("\nEarliest forecast_time:", gdf_wn['forecast_time'].min())
print("Latest forecast_time:", gdf_wn['forecast_time'].max())
print("\nLead time sample:")
print(gdf_wn[['init_time', 'forecast_time', 'lead_time_days']].head(10))

# Print unique lead times
unique_lead_times = gdf_wn['lead_time_days'].unique()
print("\nUnique lead times (in days):", unique_lead_times)

# ====================================================
# 3) Filter to a Single Lead Time (e.g., 14-day lead)
# ====================================================
# Uncomment the following line to filter for a specific lead time:
gdf_wn = gdf_wn[gdf_wn['lead_time_days'] == 14]
# Print unique lead times
unique_lead_times = gdf_wn['lead_time_days'].unique()
print("\nUnique lead times (in days):", unique_lead_times)
```

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
from shapely.geometry import Point
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------------------------------
# 1. Prepare the GeoDataFrame and spatial filtering in a projected CRS
# -----------------------------------------------------------------

# Ensure that the 'location' column exists (convert geography_text WKT to geometry)
if 'location' not in gdf_wn.columns:
    gdf_wn['location'] = gdf_wn['geography_text'].apply(wkt.loads)

# Create a GeoDataFrame in geographic CRS (EPSG:4326)
gdf_wn = gpd.GeoDataFrame(gdf_wn, geometry='location', crs='EPSG:4326')

# Reproject to EPSG:3005 (a projected CRS suitable for BC)
gdf_proj = gdf_wn.to_crs(epsg=3005)

# Define Vancouver and Lytton points in geographic CRS and project them
vancouver_point = Point(-123.12, 49.28)
lytton_point = Point(-121.95, 50.68)
vancouver_proj = gpd.GeoSeries([vancouver_point], crs='EPSG:4326').to_crs(epsg=3005).iloc[0]
lytton_proj = gpd.GeoSeries([lytton_point], crs='EPSG:4326').to_crs(epsg=3005).iloc[0]

# Set a buffer radius in meters (e.g., 10 km)
buffer_radius_m = 10000

# Create separate subsets for Vancouver and Lytton based on distance
subset_vancouver_proj = gdf_proj[gdf_proj.distance(vancouver_proj) < buffer_radius_m].copy()
subset_lytton_proj = gdf_proj[gdf_proj.distance(lytton_proj) < buffer_radius_m].copy()

# Reproject back to geographic CRS for ERA5 lookups
subset_vancouver = subset_vancouver_proj.to_crs(epsg=4326)
subset_lytton = subset_lytton_proj.to_crs(epsg=4326)

# -----------------------------------------------------------------
# 2. Preprocess forecast times and compute ERA5 temperatures
# -----------------------------------------------------------------

# Precompute tz-naive forecast times (to speed up ERA5 lookups)
subset_vancouver['forecast_time_naive'] = pd.to_datetime(subset_vancouver['forecast_time']).dt.tz_localize(None)
subset_lytton['forecast_time_naive'] = pd.to_datetime(subset_lytton['forecast_time']).dt.tz_localize(None)

# Define a function to compute ERA5 temperature at a given location and time
def compute_era5_temps(df, ds):
    era5_temps = []
    for idx, row in df.iterrows():
        pt = row['location']
        lon, lat = pt.x, pt.y
        time = row['forecast_time_naive']
        # Select nearest ERA5 temperature (in Kelvin) and convert to Celsius
        ds_time = ds.sel(valid_time=time, method="nearest")
        temp_K = float(ds_time['t2m'].sel(latitude=lat, longitude=lon, method="nearest").values)
        era5_temps.append(temp_K - 273.15)
    return era5_temps

# Compute ERA5 temperatures for each region
subset_vancouver['era5_temp_c'] = compute_era5_temps(subset_vancouver, ds_combined)
subset_lytton['era5_temp_c'] = compute_era5_temps(subset_lytton, ds_combined)

# Ensure that the WeatherNext predicted temperature column exists (convert from Kelvin if needed)
if 'temp_c' not in subset_vancouver.columns:
    subset_vancouver['temp_c'] = subset_vancouver['avg_2m_temperature'] - 273.15
if 'temp_c' not in subset_lytton.columns:
    subset_lytton['temp_c'] = subset_lytton['avg_2m_temperature'] - 273.15

# -----------------------------------------------------------------
# 3. Regression analysis for each region
# -----------------------------------------------------------------

# Vancouver regression
X_v = subset_vancouver['era5_temp_c'].values.reshape(-1, 1)
Y_v = subset_vancouver['temp_c'].values
model_v = LinearRegression()
model_v.fit(X_v, Y_v)
Y_pred_v = model_v.predict(X_v)
R2_v = model_v.score(X_v, Y_v)
RMSE_v = np.sqrt(mean_squared_error(Y_v, Y_pred_v))

print("Vancouver Regression Results:")
print("  Slope:     {:.3f}".format(model_v.coef_[0]))
print("  Intercept: {:.3f}".format(model_v.intercept_))
print("  R²:        {:.3f}".format(R2_v))
print("  RMSE:      {:.3f} °C".format(RMSE_v))

# Lytton regression
X_l = subset_lytton['era5_temp_c'].values.reshape(-1, 1)
Y_l = subset_lytton['temp_c'].values
model_l = LinearRegression()
model_l.fit(X_l, Y_l)
Y_pred_l = model_l.predict(X_l)
R2_l = model_l.score(X_l, Y_l)
RMSE_l = np.sqrt(mean_squared_error(Y_l, Y_pred_l))

print("\nLytton Regression Results:")
print("  Slope:     {:.3f}".format(model_l.coef_[0]))
print("  Intercept: {:.3f}".format(model_l.intercept_))
print("  R²:        {:.3f}".format(R2_l))
print("  RMSE:      {:.3f} °C".format(RMSE_l))

# -----------------------------------------------------------------
# 4. Plot scatter plots with fitted regression lines for both regions
# -----------------------------------------------------------------

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# --- Vancouver plot in BLUE ---
axs[0].scatter(X_v, Y_v, color='blue', alpha=0.6, label='Data Points')
# Generate fitted line for Vancouver
x_range_v = np.linspace(X_v.min(), X_v.max(), 100).reshape(-1, 1)
y_range_v = model_v.predict(x_range_v)
axs[0].plot(x_range_v, y_range_v, color='blue', linewidth=2,
            label=f'Fitted line:\ny = {model_v.coef_[0]:.3f}x + {model_v.intercept_:.3f}\n$R^2$ = {R2_v:.3f}\nRMSE = {RMSE_v:.3f}°C')
axs[0].set_xlabel('ERA5 Temperature (°C)')
axs[0].set_ylabel('WeatherNext Predicted Temperature (°C)')
axs[0].set_title('Vancouver Regression')
axs[0].legend()
axs[0].grid(True)

# --- Lytton plot in RED ---
axs[1].scatter(X_l, Y_l, color='red', alpha=0.6, label='Data Points')
# Generate fitted line for Lytton
x_range_l = np.linspace(X_l.min(), X_l.max(), 100).reshape(-1, 1)
y_range_l = model_l.predict(x_range_l)
axs[1].plot(x_range_l, y_range_l, color='red', linewidth=2,
            label=f'Fitted line:\ny = {model_l.coef_[0]:.3f}x + {model_l.intercept_:.3f}\n$R^2$ = {R2_l:.3f}\nRMSE = {RMSE_l:.3f}°C')
axs[1].set_xlabel('ERA5 Temperature (°C)')
axs[1].set_ylabel('WeatherNext Predicted Temperature (°C)')
axs[1].set_title('Lytton Regression')
axs[1].legend()
axs[1].grid(True)

plt.suptitle('Region-Specific Regression Analysis', fontsize=16)
plt
```

```python
import subprocess
import pandas as pd
import geopandas as gpd
from shapely import wkt, geometry
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# --- 1. Load New WeatherNext Data (7-Day Lead Time) ---
cmd = ['gsutil', 'ls', 'gs://gencast/weathernext/weathernext_bc_heatwave_2021_avg_7days/*.csv']
file_list = subprocess.check_output(cmd).decode('utf-8').strip().split('\n')
print("Files to load:", file_list)

# Define column names as provided
column_names = [
    "init_time",
    "forecast_time",
    "geography_polygon_text",
    "geography_text",
    "avg_total_precipitation_12hr",
    "avg_100m_u_component_of_wind",
    "avg_100m_v_component_of_wind",
    "avg_10m_u_component_of_wind",
    "avg_10m_v_component_of_wind",
    "avg_2m_temperature",
    "avg_mean_sea_level_pressure",
    "avg_sea_surface_temperature",
]

df_list = []
for f in file_list:
    df_temp = pd.read_csv(f, names=column_names, header=None)
    df_list.append(df_temp)
df_wn = pd.concat(df_list, ignore_index=True)
print("WeatherNext DataFrame shape:", df_wn.shape)
print(df_wn.head())

# --- 2. Preprocess and Describe the Data ---
# Convert time columns to datetime
df_wn['init_time'] = pd.to_datetime(df_wn['init_time'])
df_wn['forecast_time'] = pd.to_datetime(df_wn['forecast_time'])
# Convert polygon text to geometry and create a GeoDataFrame in EPSG:4326
df_wn['geometry'] = df_wn['geography_polygon_text'].apply(wkt.loads)
gdf_wn = gpd.GeoDataFrame(df_wn, geometry='geometry', crs='EPSG:4326')
# Create predicted temperature in Celsius
gdf_wn['temp_c'] = gdf_wn['avg_2m_temperature'] - 273.15

# Describe the data
print(gdf_wn.describe(include='all'))

# Quick plot for one forecast time slice
target_time_plot = gdf_wn['forecast_time'].min()
gdf_slice = gdf_wn[gdf_wn['forecast_time'] == target_time_plot]
fig, ax = plt.subplots(figsize=(10,8), subplot_kw={'projection': ccrs.PlateCarree()})
gdf_slice.plot(ax=ax, column='temp_c', cmap='RdYlBu_r', legend=True)
ax.set_title(f"WeatherNext 2m Temperature at {target_time_plot}")
ax.coastlines()
plt.show()

# --- 3. Regression Analysis for Vancouver and Lytton ---
# To reliably filter by distance, reproject to EPSG:3005 (projected CRS for BC)
gdf_proj = gdf_wn.to_crs(epsg=3005)
# Compute centroids in projected CRS
gdf_proj['centroid'] = gdf_proj.geometry.centroid

# Define Vancouver and Lytton points (in geographic CRS) and project them to EPSG:3005
vancouver_geo = geometry.Point(-123.12, 49.28)
lytton_geo = geometry.Point(-121.95, 50.68)
vancouver_proj_point = gpd.GeoSeries([vancouver_geo], crs='EPSG:4326').to_crs(epsg=3005).iloc[0]
lytton_proj_point = gpd.GeoSeries([lytton_geo], crs='EPSG:4326').to_crs(epsg=3005).iloc[0]

# Set a buffer radius in meters (e.g., 10 km)
buffer_radius_m = 10000

# Filter using the projected centroids
gdf_vancouver_proj = gdf_proj[gdf_proj['centroid'].distance(vancouver_proj_point) < buffer_radius_m].copy()
gdf_lytton_proj = gdf_proj[gdf_proj['centroid'].distance(lytton_proj_point) < buffer_radius_m].copy()

print("Vancouver subset (projected) shape:", gdf_vancouver_proj.shape)
print("Lytton subset (projected) shape:", gdf_lytton_proj.shape)

# Reproject the filtered subsets back to geographic CRS
gdf_vancouver = gdf_vancouver_proj.to_crs(epsg=4326)
gdf_lytton = gdf_lytton_proj.to_crs(epsg=4326)
# Also, reproject the centroids
gdf_vancouver['centroid'] = gdf_vancouver_proj['centroid'].to_crs(epsg=4326)
gdf_lytton['centroid'] = gdf_lytton_proj['centroid'].to_crs(epsg=4326)

print("Vancouver subset (geographic) shape:", gdf_vancouver.shape)
print("Lytton subset (geographic) shape:", gdf_lytton.shape)

# Define a function to sample ERA5 temperature at a given location and forecast time.
# (Assume ds_combined, the ERA5 dataset, is already loaded.)
def sample_era5_at_point(time, lon, lat):
    time_naive = pd.to_datetime(time).tz_localize(None)
    ds_time = ds_combined.sel(valid_time=time_naive, method="nearest")
    t2m_val = float(ds_time['t2m'].sel(latitude=lat, longitude=lon, method="nearest").values)
    return t2m_val - 273.15

# Compute ERA5 observed temperature for each record using its centroid
def compute_era5_for_region(gdf):
    era5_temps = []
    for idx, row in gdf.iterrows():
        cen = row['centroid']
        t_obs = sample_era5_at_point(row['forecast_time'], cen.x, cen.y)
        era5_temps.append(t_obs)
    gdf['era5_temp_c'] = era5_temps
    return gdf

gdf_vancouver = compute_era5_for_region(gdf_vancouver)
gdf_lytton = compute_era5_for_region(gdf_lytton)

# Regression: WeatherNext predicted temperature (Y) vs ERA5 observed temperature (X)
# --- Vancouver Regression ---
X_v = gdf_vancouver['era5_temp_c'].values.reshape(-1, 1)
Y_v = gdf_vancouver['temp_c'].values
model_v = LinearRegression()
model_v.fit(X_v, Y_v)
Y_pred_v = model_v.predict(X_v)
R2_v = model_v.score(X_v, Y_v)
RMSE_v = np.sqrt(mean_squared_error(Y_v, Y_pred_v))

print("Vancouver Regression Results:")
print("  Slope:", model_v.coef_[0])
print("  Intercept:", model_v.intercept_)
print("  R²:", R2_v)
print("  RMSE:", RMSE_v)

# --- Lytton Regression ---
X_l = gdf_lytton['era5_temp_c'].values.reshape(-1, 1)
Y_l = gdf_lytton['temp_c'].values
model_l = LinearRegression()
model_l.fit(X_l, Y_l)
Y_pred_l = model_l.predict(X_l)
R2_l = model_l.score(X_l, Y_l)
RMSE_l = np.sqrt(mean_squared_error(Y_l, Y_pred_l))

print("Lytton Regression Results:")
print("  Slope:", model_l.coef_[0])
print("  Intercept:", model_l.intercept_)
print("  R²:", R2_l)
print("  RMSE:", RMSE_l)

# --- 4. Plot Regression Scatter Plots with Fitted Lines ---
fig, axs = plt.subplots(1, 2, figsize=(14,6))

# Vancouver plot in blue
axs[0].scatter(X_v, Y_v, color='blue', alpha=0.6, label='Data points')
x_range_v = np.linspace(X_v.min(), X_v.max(), 100).reshape(-1, 1)
axs[0].plot(x_range_v, model_v.predict(x_range_v), color='blue', linewidth=2,
            label=f'Fitted: y = {model_v.coef_[0]:.3f}x + {model_v.intercept_:.3f}\n$R^2$ = {R2_v:.3f}\nRMSE = {RMSE_v:.3f}')
axs[0].set_xlabel('ERA5 Temperature (°C)')
axs[0].set_ylabel('WeatherNext Predicted Temperature (°C)')
axs[0].set_title('Vancouver Regression')
axs[0].legend()
axs[0].grid(True)

# Lytton plot in red
axs[1].scatter(X_l, Y_l, color='red', alpha=0.6, label='Data points')
x_range_l = np.linspace(X_l.min(), X_l.max(), 100).reshape(-1, 1)
axs[1].plot(x_range_l, model_l.predict(x_range_l), color='red', linewidth=2,
            label=f'Fitted: y = {model_l.coef_[0]:.3f}x + {model_l.intercept_:.3f}\n$R^2$ = {R2_l:.3f}\nRMSE = {RMSE_l:.3f}')
axs[1].set_xlabel('ERA5 Temperature (°C)')
axs[1].set_ylabel('WeatherNext Predicted Temperature (°C)')
axs[1].set_title('Lytton Regression')
axs[1].legend()
axs[1].grid(True)

plt.suptitle('Regression Analysis: 7-Day Lead Time WeatherNext vs ERA5', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
from shapely import wkt, geometry
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# --- ASSUMPTION: 
# gdf_wn is the GeoDataFrame loaded from your 1-day lead time data 
# (with columns: init_time, forecast_time, geography_polygon_text, geography_text, avg_2m_temperature, etc.)
# and ds_combined is your ERA5 reanalysis dataset.

# Ensure predicted temperature in Celsius exists
if 'temp_c' not in gdf_wn.columns:
    gdf_wn['temp_c'] = gdf_wn['avg_2m_temperature'] - 273.15

# Create a "location" column from geography_text if needed (which holds a POINT)
if 'location' not in gdf_wn.columns:
    gdf_wn['location'] = gdf_wn['geography_text'].apply(wkt.loads)

# --- Spatial Filtering for Vancouver and Lytton ---
# Reproject to a projected CRS (EPSG:3005) for accurate distance calculations
gdf_proj = gdf_wn.to_crs(epsg=3005)
# Compute centroids (here we use the polygon geometry's centroid)
gdf_proj['centroid'] = gdf_proj.geometry.centroid

# Define Vancouver and Lytton points (geographic) and project them to EPSG:3005
vancouver_geo = geometry.Point(-123.12, 49.28)
lytton_geo = geometry.Point(-121.95, 50.68)
vancouver_proj_point = gpd.GeoSeries([vancouver_geo], crs='EPSG:4326').to_crs(epsg=3005).iloc[0]
lytton_proj_point = gpd.GeoSeries([lytton_geo], crs='EPSG:4326').to_crs(epsg=3005).iloc[0]

# Use a 10 km buffer (10,000 meters)
buffer_radius_m = 10000

# Filter polygons whose centroids lie within the buffer of each region
gdf_vancouver_proj = gdf_proj[gdf_proj['centroid'].distance(vancouver_proj_point) < buffer_radius_m].copy()
gdf_lytton_proj = gdf_proj[gdf_proj['centroid'].distance(lytton_proj_point) < buffer_radius_m].copy()

# Reproject the filtered subsets back to geographic CRS
gdf_vancouver = gdf_vancouver_proj.to_crs(epsg=4326)
gdf_lytton = gdf_lytton_proj.to_crs(epsg=4326)
# Also update centroids (in geographic CRS) for ERA5 sampling
gdf_vancouver['centroid'] = gdf_vancouver_proj['centroid'].to_crs(epsg=4326)
gdf_lytton['centroid'] = gdf_lytton_proj['centroid'].to_crs(epsg=4326)

# --- ERA5 Sampling Function ---
def sample_era5_at_point(time, lon, lat):
    # Convert forecast time to tz-naive to match ds_combined
    time_naive = pd.to_datetime(time).tz_localize(None)
    ds_time = ds_combined.sel(valid_time=time_naive, method="nearest")
    t2m_val = float(ds_time['t2m'].sel(latitude=lat, longitude=lon, method="nearest").values)
    return t2m_val - 273.15  # Convert from Kelvin to Celsius

def compute_era5_for_region(gdf):
    era5_temps = []
    for idx, row in gdf.iterrows():
        cen = row['centroid']
        t_obs = sample_era5_at_point(row['forecast_time'], cen.x, cen.y)
        era5_temps.append(t_obs)
    gdf['era5_temp_c'] = era5_temps
    return gdf

# Compute ERA5 temperatures for each region
gdf_vancouver = compute_era5_for_region(gdf_vancouver)
gdf_lytton = compute_era5_for_region(gdf_lytton)

# --- Regression Analysis ---
# For Vancouver
X_v = gdf_vancouver['era5_temp_c'].values.reshape(-1, 1)
Y_v = gdf_vancouver['temp_c'].values
model_v = LinearRegression()
model_v.fit(X_v, Y_v)
Y_pred_v = model_v.predict(X_v)
R2_v = model_v.score(X_v, Y_v)
RMSE_v = np.sqrt(mean_squared_error(Y_v, Y_pred_v))

# For Lytton
X_l = gdf_lytton['era5_temp_c'].values.reshape(-1, 1)
Y_l = gdf_lytton['temp_c'].values
model_l = LinearRegression()
model_l.fit(X_l, Y_l)
Y_pred_l = model_l.predict(X_l)
R2_l = model_l.score(X_l, Y_l)
RMSE_l = np.sqrt(mean_squared_error(Y_l, Y_pred_l))

print("Vancouver Regression Results:")
print("  Slope: {:.3f}".format(model_v.coef_[0]))
print("  Intercept: {:.3f}".format(model_v.intercept_))
print("  R²: {:.3f}".format(R2_v))
print("  RMSE: {:.3f} °C".format(RMSE_v))

print("\nLytton Regression Results:")
print("  Slope: {:.3f}".format(model_l.coef_[0]))
print("  Intercept: {:.3f}".format(model_l.intercept_))
print("  R²: {:.3f}".format(R2_l))
print("  RMSE: {:.3f} °C".format(RMSE_l))

# --- Plotting the Regression Scatter Plots ---
fig, axs = plt.subplots(1, 2, figsize=(14,6))

# Vancouver Plot (blue)
axs[0].scatter(X_v, Y_v, color='blue', alpha=0.6, label='Data points')
x_range_v = np.linspace(X_v.min(), X_v.max(), 100).reshape(-1, 1)
axs[0].plot(x_range_v, model_v.predict(x_range_v), color='blue', linewidth=2,
            label=f'Fitted: y={model_v.coef_[0]:.3f}x+{model_v.intercept_:.3f}\n$R^2$={R2_v:.3f}, RMSE={RMSE_v:.3f}')
axs[0].set_xlabel('ERA5 Temperature (°C)')
axs[0].set_ylabel('WeatherNext Temperature (°C)')
axs[0].set_title('Vancouver Regression')
axs[0].legend()
axs[0].grid(True)

# Lytton Plot (red)
axs[1].scatter(X_l, Y_l, color='red', alpha=0.6, label='Data points')
x_range_l = np.linspace(X_l.min(), X_l.max(), 100).reshape(-1, 1)
axs[1].plot(x_range_l, model_l.predict(x_range_l), color='red', linewidth=2,
            label=f'Fitted: y={model_l.coef_[0]:.3f}x+{model_l.intercept_:.3f}\n$R^2$={R2_l:.3f}, RMSE={RMSE_l:.3f}')
axs[1].set_xlabel('ERA5 Temperature (°C)')
axs[1].set_ylabel('WeatherNext Temperature (°C)')
axs[1].set_title('Lytton Regression')
axs[1].legend()
axs[1].grid(True)

plt.suptitle('Regression Analysis: 1-Day Lead Time WeatherNext vs ERA5', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
